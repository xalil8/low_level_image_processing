```python
import numpy as np
import PIL as Image
import cv2 
from matplotlib import pyplot as plt
```

created empty rgb arrays with specific sizes to test  padding 

```python
h,w= 18,18
a = np.ones((h,w,3),dtype=np.uint8)
b = np.ones((h,w,3),dtype=np.uint8)
c = np.ones((h,w,3),dtype=np.uint8)
a = a*22
b = b*100
c = c*200
a[9][9] *= 255
b[9][9] *= 255
c[9][9] *= 0
```

```python
#plt.imshow(a, interpolation='nearest',extent=(0, 5, 0, 8))
#plt.imshow(c, interpolation='nearest')
#plt.show()

f, axarr = plt.subplots(nrows=1,ncols=3)
plt.sca(axarr[0]); 
plt.imshow(a); plt.title('title 1')
plt.sca(axarr[1]); 
plt.imshow(b); plt.title('title 2')
plt.sca(axarr[2]); 
plt.imshow(c); plt.title('title 3')
plt.show()
```

how to add padding to rgb single image

first tuple(h_first, height_last), (width_first, width_last) (chn,chn), only adjust first 2 one 

```python
padded = np.pad(a, ((3, 3), (1, 2), (0, 0)), 'constant', constant_values=255)    
plt.imshow(padded, interpolation='nearest')
plt.show()
```


stacked 3 image into an numpay array
